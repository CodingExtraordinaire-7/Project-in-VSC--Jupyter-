```python
#import statements

import pandas as pd
```

```python
#Declaring of link 
url = "https://github.com/frontiertechinstitute/datasets/raw/refs/heads/main/Passwords/sampled_passwords.csv"
pw = pd.read_csv(url)

pw.info()
```

```python
pw = pw.rename(columns={"Strength_Level": "Strength Level"})

new_pw = pw.drop_duplicates()
new_pw.info() #checks to make sure that rows were actually dropped
new_pw.isnull().sum() #checks to see if there are any missing values in the dataset
```

```python
new_pw['Password Length'] = new_pw['Password'].str.len() #Makes a new column for password character length

#function made to calculate how many times password characters repeat, retunrs a percentage of unique characters
def password_diversity(password):
    unique_chars = len(set(password))
    total_chars = len(password)
    result = unique_chars / total_chars * 100

    return round(result)

new_pw['Password Diversity'] = new_pw['Password'].apply(password_diversity)

new_pw.info()
```

```python
#predictability function, return a percentage on how predicatable a password is

def predictability(password):

    predict_score = 0

    #First condition: if the password starts with a capital letter (+25 pts)
    first_char = password[0]
    if first_char.isupper():
        predict_score += 25 #reward

    #Second condition: if the password ends with a special character (+25 pts)
    last_char = password[-1]
    if last_char.isalpha() and not last_char.isdigit():
        predict_score += 25 #reward

    #Third condition: if the password contains a repeating number (25+ pts)
    #goes through password and saves each character (in groups of 3)
    for i in range(len(password)-1): 
        char1 = password[i]
        char2 = password[i + 1]

        if char1 == char2: #checks if all 3 characters equal each other
            predict_score += 25 #reward
            break #stops the for loop if a repeating set is found

    #Fourth condition: if the password ends with a number
    if last_char.isdigit():
        predict_score += 25 #reward

    return predict_score

new_pw["Predictability Score (%)"] = new_pw["Password"].apply(predictability)
```
